```python
from sidekit.features_extractor import plp
from pydub.utils import mediainfo
import matplotlib.pyplot as plt
import librosa.display
import numpy as np
import librosa
```

```python
%matplotlib inline
```

```python
audio_file_path = r"C:\Users\admin\Documents\AI\data\voice_recognition\train\3\0.wav"
sample_rate = int(mediainfo(audio_file_path)['sample_rate'])
```

```python
y_mel, sr_mel = librosa.load(path=audio_file_path, sr=sample_rate)
y_rasta_plp, sr_rasta_plp = librosa.load(path=audio_file_path, sr=sample_rate)
y_plp, sr_plp = librosa.load(path=audio_file_path, sr=sample_rate)

print("y_mel shape: ", np.shape(y_mel))
print("y_rasta_plp shape: ", np.shape(y_rasta_plp))
print("y_plp shape: ", np.shape(y_plp))
```

```python
plt.plot(y_mel)
plt.title('y_mel')
plt.show()
print('max_value: ', np.max(y_mel))
print('min_value: ', np.min(y_mel))

plt.plot(y_rasta_plp)
plt.title('y_rasta_plp')
plt.show()
print('max_value: ', np.max(y_rasta_plp))
print('min_value: ', np.min(y_rasta_plp))

plt.plot(y_plp)
plt.title('y_plp')
plt.show()
print('max_value: ', np.max(y_plp))
print('min_value: ', np.min(y_plp))
```

```python
y_mel = y_mel * 10
y_rasta_plp = y_rasta_plp * 10
y_plp = y_plp * 10
```

```python
s_mel = librosa.feature.melspectrogram(y=y_mel, sr=sr_mel)
s_log_mel = librosa.amplitude_to_db(S=s_mel)

librosa.display.specshow(librosa.power_to_db(s_mel, ref=np.max), sr=sr_mel, x_axis='time', y_axis='mel')
plt.title('mel power spectrogram')
# plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
plt.show()

min_level_db = -100

def _normalize(s):
    return np.clip((s - min_level_db) / -min_level_db, 0, 1)

norm_melspectro = _normalize(s_log_mel)
librosa.display.specshow(data=norm_melspectro, sr=sr_mel)
plt.title('mel power spectrogram')
# plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
plt.show()
```

```python
s_rasta_plp = plp(y_rasta_plp, fs=sr_rasta_plp)

librosa.display.specshow(data=s_rasta_plp[0], sr=sr_rasta_plp)
plt.title('rasta_plp')
# plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
plt.show()

print("shape of s_rasta_plp: ", np.shape(s_rasta_plp[0]))
print("max value in s_rasta_plp[0]: ", np.max(s_rasta_plp[0]))
print("min value in s_rasta_plp[0]: ", np.min(s_rasta_plp[0]))
```

```python
s_plp = plp(y_plp, fs=sr_plp, rasta=False)

librosa.display.specshow(data=s_plp[0], sr=sr_plp)
plt.title('plp')
# plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
plt.show()

print("shape of s_plp: ", np.shape(s_plp[0]))
print("max value in s_plp[0]: ", np.max(s_plp[0]))
print("min value in s_plp[0]: ", np.min(s_plp[0]))
```
